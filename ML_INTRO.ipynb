### **SOME DATA PREPROCESSING WITH SCIKIT-LEARN**
The scikit-learn library provides several ready to use method to perform data preprocessing [link](https://https://scikit-learn.org/stable/modules/preprocessing.html)

**How** to normalize data between 0 and 1 - [link](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.MinMaxScaler.html)

```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler
X_example = np.array([[ 1., -1.,  2.],
                     [ 2.,  0.,  0.],
                     [ 0.,  1., -1.]])

min_max_scaler = MinMaxScaler()

X_example_minmax = min_max_scaler.fit_transform(X_example)

print(X_example_minmax)
```

How to normalize data between 0 and 1 - [link](https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.StandardScaler.html)

```python
from sklearn.preprocessing import StandardScaler

X_example = np.array([[ 1., -1.,  2.],
                    [ 2.,  0.,  0.],
                    [ 0.,  1., -1.]])

standard_scaler = StandardScaler()

X_example_minmax = standard_scaler.fit_transform(X_example)

print(X_example_minmax)
```

## Loading again the Iris dataset

```python
from sklearn.datasets import load_iris
import numpy as np

iris = load_iris()

# store feature matrix in "X"
X = iris.data

# store response vector in "y"
y = iris.target
```

### **Divide the data in training and test set**

Import the *train_test_split* function to randomly divide the dataset in training and test

```python
from sklearn.model_selection import train_test_split
```

Deploy the *train_test_split* function on the iris data to partition the original data with a proportion of 60% of the data for the training set and 40% of the data for the test set

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.40)
```

```python
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

### K-nearest neighbors (KNN) classification
* Pick a value for K.
* Search for the K observations in the training data that are "nearest" to the measurements of the unknown iris.
* Use the most popular response value from the K nearest neighbors as the predicted response value for the unknown iris.

### **COMPUTE EVALUATION METRICS TO ASSESS THE MODEL PERFORMANCES**

The evaluation metrics available in Scikit-Learn are listed and discussed at this [link](https://scikit-learn.org/stable/modules/model_evaluation.html)

Import the Accuracy and the F-Score metrics

```python
from sklearn.metrics import f1_score, accuracy_score
```

Compute the accuracy achieved by the KNN algorithm on the test set - [link](https://scikit-learn.org/stable/modules/generated/sklearn.metrics.accuracy_score.html)

```python
accuracy = accuracy_score(y_test, pred)
print(accuracy)
```

Compute the F1-score achieved by the KNN algorithm on the test set via a weighted average (the weight is the number of examples of each class in the test set) - [link](https://scikit-learn.org/stable/modules/generated/sklearn.metrics.f1_score.html)

```python
f_score = f1_score(y_test, pred, average='weighted')
print(f_score)
```

Compute the F1-score achieved by the KNN algorithm on the test set per class

```python
f_scores = f1_score(y_test, pred, average=None)
print(f_scores)
```

Visualize the per class performances with bar chart via matplotlib - [link](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.bar.html)

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

#fruits = ['apple', 'blueberry', 'cherry', 'orange']
#counts = [40, 100, 30, 55]
#bar_labels = ['red', 'blue', '_red', 'orange']
bar_colors = ['tab:red', 'tab:blue', 'tab:orange']

ax.bar(iris.target_names, f_scores, label=iris.target_names)#, color=bar_colors)

ax.set_ylabel('F-Score')
ax.set_title('Classes')
#ax.legend(title='')

plt.show()
```

Compute the confusion matrix between the KNN algorithm prediction and the real test set classes - [link](https://scikit-learn.org/stable/modules/generated/sklearn.metrics.confusion_matrix.html)

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, pred, normalize='true')
print(cm)
```

Visualize the confusion matrix via a Heat Map with matplotlib - [link](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.imshow.html)

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
#im = ax.imshow(cm,cmap="seismic")
im = plt.imshow(cm,cmap="seismic")

# Show all ticks and label them with the respective list entries
ax.set_xticks(np.arange(len(iris.target_names)), labels=iris.target_names)
ax.set_yticks(np.arange(len(iris.target_names)), labels=iris.target_names)

plt.setp(ax.get_xticklabels(), rotation=45, ha="right",
         rotation_mode="anchor")

for i in range(len(iris.target_names)):
    for j in range(len(iris.target_names)):
        val = int(cm[i,j] * 1000) / 1000.0
        text = ax.text(j, i, val,
                       ha="center", va="center", color="w")

plt.colorbar()
plt.show()
```

### **PERFORM AN ASSESSMENT WITH K-FOLD CROSS VALIDATION (K=5)**

The Cross Validation procedure allows to estimate the error performances of the model considering each sample once, and only once, as test sample - [link](https://scikit-learn.org/stable/modules/cross_validation.html)

Import the function that implements the cross validation data splitting in scikit-learn - [link](https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.KFold.html)

```python
from sklearn.model_selection import KFold
```

Load again the Iris Dataset

```python
from sklearn.datasets import load_iris
import numpy as np

iris = load_iris()

# store feature matrix in "X"
X = iris.data

# store response vector in "y"
y = iris.target
```

* We set the number of fold equals to 5
* For each fold, the kf.split() function returns the data indexes to partition the data in training and test
* We train a knn model with the training data
* We apply the model on the specific test data
* We store the current performances in a vector (acc_values)
* We finally compute the average and standard deviation of the 5 accuracies we have previously obtain



```python
num_splits = 5
#kf = KFold(n_splits=num_splits, shuffle=True)
kf = KFold(n_splits=num_splits)


acc_values = []
for train_idx, test_idx in kf.split(X):
  knn = KNeighborsClassifier(n_neighbors=1)
  temp_x_train = X[train_idx]
  temp_y_train = y[train_idx]

  temp_x_test = X[test_idx]
  temp_y_test = y[test_idx]
  knn.fit(temp_x_train, temp_y_train)
  pred = knn.predict(temp_x_test)
  temp_acc = accuracy_score(temp_y_test, pred)
  acc_values.append( temp_acc )

print("Average with standard deviation accuracy %.4f +- %.4f"% (np.mean(acc_values), np.std(acc_values))  )

```

In Scikit-Learn we can compute directly the Cross Validation score with the function cross_val_score - [link](https://scikit-learn.org/stable/modules/generated/sklearn.model_selection.cross_val_score.html)

```python
from sklearn.model_selection import cross_val_score

num_splits = 20
knn = KNeighborsClassifier(n_neighbors=1)
acc_values = cross_val_score(knn, X, y, cv=num_splits)
print(acc_values)
print("Average with standard deviation accuracy %.4f +- %.4f"% (np.mean(acc_values), np.std(acc_values))  )
```

Visualize the average and standard deviation with a box plot with matplotlib - [link](https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.boxplot.html)

```python
fig1, ax1 = plt.subplots()
ax1.set_title('Accuracy Box Plot')
ax1.boxplot(acc_values)

plt.show()
```
